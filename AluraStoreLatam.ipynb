### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 4. Productos más y menos vendidos

```python
# Conteo de productos por tienda
ventas_por_producto_1 = tienda.groupby(['Lugar de Compra','Producto']).size().reset_index(name='Cantidad')

# Ordenar por cantidad dentro de cada tienda
ventas_ordenadas_1 = ventas_por_producto_1.sort_values(['Lugar de Compra','Cantidad'], ascending=[True, False])

# Extraer el producto más vendido por cada tienda
producto_mas_vendido_1 = ventas_ordenadas_1.groupby('Lugar de Compra').head(1)

print(producto_mas_vendido_1)
```

```python
# Conteo de productos por tienda
ventas_por_producto_2 = tienda2.groupby(['Lugar de Compra','Producto']).size().reset_index(name='Cantidad')

# Ordenar dentro de cada tienda
ventas_ordenadas_2 = ventas_por_producto_2.sort_values(['Lugar de Compra','Cantidad'], ascending=[True, False])

# Extraer el producto más vendido por cada tienda
producto_mas_vendido_2 = ventas_ordenadas_2.groupby('Lugar de Compra').head(1)

print(producto_mas_vendido_2)
```

```python
# Conteo de productos por tienda
ventas_por_producto_3 = tienda3.groupby(['Lugar de Compra','Producto']).size().reset_index(name='Cantidad')

# Ordenar dentro de cada tienda
ventas_ordenadas_3 = ventas_por_producto_3.sort_values(['Lugar de Compra','Cantidad'], ascending=[True, False])

# Extraer el producto más vendido por cada tienda
producto_mas_vendido_3 = ventas_ordenadas_3.groupby('Lugar de Compra').head(1)

print(producto_mas_vendido_3)
```

```python
# Conteo de productos por tienda
ventas_por_producto_4 = tienda4.groupby(['Lugar de Compra','Producto']).size().reset_index(name='Cantidad')

# Ordenar dentro de cada tienda
ventas_ordenadas_4 = ventas_por_producto_4.sort_values(['Lugar de Compra','Cantidad'], ascending=[True, False])

# Extraer el producto más vendido por cada tienda
producto_mas_vendido_4 = ventas_ordenadas_4.groupby('Lugar de Compra').head(1)

print(producto_mas_vendido_4)
```
